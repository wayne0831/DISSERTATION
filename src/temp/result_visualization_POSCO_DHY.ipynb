```python
##################################################################################################################################################################
# set result directory
##################################################################################################################################################################

res_dir_pred = RES_PATH['PRED_ROOT'] + RES_PATH['CDDA_DHY']
res_dir_perf = RES_PATH['PERF_ROOT'] + RES_PATH['CDDA_DHY']

file_name    = '250208_REF_REC_APP_POSCO_XGB_WSTD_REC'
```

```python
res_df_pred_v1 = pd.read_csv(f'{res_dir_pred}{file_name}_V1_PRED.csv')
res_df_perf_v1 = pd.read_csv(f'{res_dir_perf}{file_name}_V1_PERF.csv')

res_df_perf = res_df_perf_v1
res_df_pred = res_df_pred_v1
target      = '442T05_Li'
init_tr_end_idx  = res_df_perf['init_tr_end_idx'][0]
mdl_upd_idx_list = [sublist[-1] for sublist in eval(res_df_perf['adapt_prd'][0])]
mdl_upd_idx_list = [idx for idx in mdl_upd_idx_list if idx != 'NONE']

res_df_pred['time'] = pd.to_datetime(res_df_pred['time'])

plt.figure(figsize=(6, 4))
plt.plot(res_df_pred['time'], res_df_pred['y_real'], label='Real', color='black')
plt.plot(res_df_pred['time'], res_df_pred['y_pred'], label='Pred', color='red')
plt.legend()
plt.ylabel(target)
for idx in mdl_upd_idx_list:
    plt.axvline(x=res_df_pred['time'][idx-init_tr_end_idx], linestyle='--')
# end for
plt.show()
```
